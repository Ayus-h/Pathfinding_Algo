```python
import numpy as np
B = tuple([int(i) for i in input("Enter your begining vertex ").split(",")])
E = tuple([int(i) for i in input("Enter your end vertex ").split(",")])
grid = np.zeros((10,10)).astype(int)

grid[B] = "-1"
grid[E] = "-2"

def neighbour(grid,pos):
    x1,y1,x2,y2 = limit(grid,*pos)
    x1,y1,x2,y2 = limit(grid,*pos)
    index,val = [],[]
    for x in range(x1, x2):
        for y in range(y1, y2):
            if ((x,y) != pos and (x,y) != "" and grid[x,y] ==0) or ((x,y) != pos and grid[x,y] == -2):
                index.append((x,y))
                val.append(grid[x,y])
            elif (x,y) == pos:
                cval = grid[x,y]
    return index,val,cval
    
def limit(grid,x,y):
    x1,x2,y1,y2 = x-1,x+2,y-1,y+2
    x_max = grid.shape[0]
    y_max = grid.shape[1]
    if x1<0:
        x1= 0
    if x2> x_max:
        x2 = x_max
    if y1<0:
        y1= 0
    if y2> y_max:
        y2 = y_max
    return x1,y1,x2,y2
    
def adder(grid,pos):
    index,val,cval = neighbour(grid,pos)
    if cval<0:
        cval = 0
    for i in range(len(val)):
        if val[i] == 0:
            grid[index[i]] = cval+1
    return index,val
    
def pathfinder(grid,pos):
        value = [1]
        while -2 not in value:
            li = []
            value = []
            for i in pos:
                ind,val = adder(grid,i)
                li +=ind
                value+=val
                if -2 in value:
                    break
            pos = li
  

grid[0:6,5] = -3
        
print(grid,end="\n\n")
```

```python
Origin_grid = grid.copy()
pos = [B]
pathfinder(grid,pos)
print(grid)
```

```python
def backtrack(grid,pos):
    x1,y1,x2,y2 = limit(grid,*pos)
    index,val = [],[]
    for x in range(x1,x2):
        for y in range(y1,y2):
            if (x,y) != pos and grid[x,y] !=0 and grid[x,y] != -2 and grid[x,y] != -3:
                index.append((x,y))
                val.append(grid[x,y])
    return index,val
   
```

```python
def findingPath(grid,pos):
    if grid[pos] == -1:
        return 0
    index,val = backtrack(grid,pos)
    min_v = min(val)
    i = val.index(min_v)
    pos = index[i]
    print(pos,min_v)
    Origin_grid[pos] = min_v
    findingPath(grid,pos)
```
